```python
import pandas as pd

df = pd.read_csv(
    "../Subway_base_network/seoul.txt",
    delimiter=" ",
    names=["no", "name", "line"],
    encoding="UTF-8",
)

# only nodes
df = df.iloc[
    :705,
]


def no_to_name(no):
    """
    parameter: stNO
    return: stName, line
    """
    name = df[df["no"] == no]["name"].values[0]
    line = df[df["no"] == no]["line"].values[0]
    return name, line
def name_to_no(name, line):
    x = df[df["name"] == name]
    y = x[x['line'] == line]
    return y['no'].values[0]
name_to_no("사당", '2')
```
