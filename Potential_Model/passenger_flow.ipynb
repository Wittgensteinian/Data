```python
data = pd.read_csv("data.csv")
line_convert_table = {"1호선": "1", "2호선": "2", "3호선": "3", "4호선": "4", "5호선": "5", "6호선": "6", "7호선": "7", "8호선": "8", "9호선": "9",
                     "중앙선": "K1", '우이신설선': 'S1', '분당선': "K2", '수인선': 'K2_'}
"""
위의 테이블은 혹시 하나의 역도 역번호가 다르면 다른 것으로 처리할까 해서
만들다 만 테이블인데, 역이름이 같으면 같은 역으로 보기로 합의했기에 필요가 없어짐
일단은 지우지 않고 냅둠
"""


for i, j in line_convert_table.items():
    data = data.replace(i, j)
print(data)
sums = data.sum(axis = 1)
```

```python
name_convert_table = {'가천대': '가천대역', '신천': '잠실새내', '이수': '총신대입구(이수)', '총신대입구': '총신대입구(이수)', '총신대입구(이수)': '총신대입구(이수)', '양평(5호선)': '양평(5호선)', '양평(중앙선)': '양평(중앙선)'}
"""
name_convert_table로 구조상 수정이 힘든건 data.csv 직접 수정
나머지는 다른 시간으로 데이터를 바꿀 것을 미리 대비해서 data.csv를 수정하지 않고 name_convert_table을 이용함



+ 2021.4월 기준으로, 뭐였는지 기억은 안 나는데 하나 새로 신설되서 지하철 네트워크에 없는 역이 있음.

데이터셋 자체를 좀 고치기도 했는데,
원래 데이셋은 2015년 1월달의 '같은 기간'에 대한 데이터의 이수역에 대해서도 7호선은 '이수', 4호선은 '총신대입구'로 저장되어 있음. 뭐지;
동두천 중앙 -> 동두천중앙
양평 -> 양평(5호선)
양평 -> 양평(중앙선)
총신대입구 -> 총신대입구(이수)
이수 -> 총신대입구(이수)


#### 
2015년 1월달 데이터는 '인천국제공항역'이 있음
근데 지금은 이게 '인천공학1터미널역'으로 바뀌고 '인천공항2터미널역' 신설(참고: https://ko.wikipedia.org/wiki/%EC%9D%B8%EC%B2%9C%EA%B5%AD%EC%A0%9C%EA%B3%B5%ED%95%AD%EC%B2%A0%EB%8F%84)

2015년 1월 대신 2017년 10월 데이터 채택

#### 
2017.10월
총신대입구(이수)로 잘 기입되어 있음
양평 -> 양평(5호선) (파일 수정)
양평 -> 양평(중앙선) (파일 수정)
7호선 이수 -> 총신대입구(이수) (name_convert_table 이용)

놀랍게도 양평은 둘이 다른 역인데 이름만 같음(...)

왜인지 인천국제공항역이 그대로 있어서 위키피디아 인천국제공항철도 문서 다시 읽어보니 '인천공항제1터미널역'으로 바꾸는 건 2018년 1월부터 적용이라고 함
2017년 6월 29일에 변경했다고 해서 2017년 10월 데이터 읽어왔는데 적용은 2018년 1월 13일이라고 한다.
2018년 3월 데이터 채택


#### 
2018.3월
양평 -> 양평(5호선) (파일 수정)
양평 -> 양평(중앙선) (파일 수정)
K2의 가천대 -> 가천대역 (name_convert_table 이용)
지평: 지하철 네트워크 데이터셋에는 없는데 왜인지 여기엔 있음.. K138역번호, 일단은 무시함.
"""
import re
name_to_flow = {}
for i in range(587):
    x = data.iloc[i]
    name = name_convert_table.get(x['지하철역'], re.sub('\(.*?\)', '', x['지하철역']))
    if name in ('지평'):
        continue
    print(name, "/", x['호선명'])
    assert(hasName(name))
    name_to_flow[name] = name_to_flow.get(name, 0) + sums[i]
```

```python
sum_flow = 0
for i, j in name_to_flow.items():
    sum_flow += int(j)
print(sum_flow)
## 아마 sum_flow는 들어오는 인원과 나가는 인원을 모두 합하므로 한 사람이 대충 2번 세졌을 것임. 일단 무시하고 진행함
for i in name_to_flow:
    name_to_flow[i] = name_to_flow[i] / sum_flow * len(name_to_flow)
print(name_to_flow)
```
